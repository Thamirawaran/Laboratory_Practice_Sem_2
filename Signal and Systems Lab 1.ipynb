#1.1 Real CT Sinusoid

```python
import numpy as np # import numpy to handle numerics
import matplotlib.pyplot as plt # import matplotlib to hadle figures
%matplotlib inline

fs = 44100 # 44,100−Hz sampling frequency
ts = 1/fs
t = np.arange(-1., 1., ts) # A linearly−spaced array with step ts
A = 1.
f = 2 # 2 Hz
w0 = 2*np.pi*f
phi = 0.
xt = A*np.cos(w0*t + phi)
fig, ax = plt.subplots()
ax.plot(t,xt)
plt.show()
```

#1.2 Real CT Exponential 

```python
fs = 4.4e4 # sampling frequency
ts = 1/fs
t = np.arange(-1., 1., ts)
C=1.0
alpha=1.2
xt=C*np.exp(alpha*t)
fig, ax = plt.subplots()
ax.plot(t,xt)
ax.set_xlabel('$t$')
ax.set_ylabel('$xt$')
ax.grid(True)
ax.title.set_text('$e^{1.2t}$')
plt.show()
```

#1.3 Growing Sinusoidal Signal

```python
fs = 4.4e4 # sampling frequency
ts = 1/fs
t = np.arange(-1., 1., ts)
f=5 #5Hz
w0=2*np.pi*f
C=0.15
r=2
phi=0
xt1=C*np.exp(r*t)*np.cos(w0*t+phi)
fig, ax1 = plt.subplots()
ax1.plot(t,xt1)
ax1.set_xlabel('$t$')
ax1.set_ylabel('$xt1$')
ax1.grid(True)
ax1.title.set_text('$0.15 e^{2t} \cos(10\pi t)$')

C=0.15
r=-2
phi=np.pi/2
xt2=C*np.exp(r*t)*np.cos(w0*t+phi)
fig, ax2 = plt.subplots()
ax2.plot(t,xt2)
ax2.set_xlabel('$t$')
ax2.set_ylabel('$xt2$')
ax2.grid(True)
ax2.title.set_text('$0.15 e^{-2t} \cos(10\pi t +\pi/2)$')
plt.show()
```

#1.7 Transformation of the Independent Variable

```python
def x(t):
  if (t < 0.):
    return 0.
  elif (t < 1.):
    return 1.
  elif (t < 2.):
    return 2. - t
  else:
    return 0.
fs = 4.4e4 # 44,000−Hz sampling frequency
ts = 1/fs
t = np.arange(-3.5, 3.5, ts) # A linearly−spaced array with step ts
fig, axes = plt.subplots(7,1, sharey='all', figsize=(10,15))
# x(t)
axes[0].plot(t, [x(t_) for t_ in t])
axes[0].set_xlabel('$t$')
axes[0].set_ylabel('$x(t)$')
axes[0].grid(True)
# Your code goes here
# x(t-1)
axes[1].plot(t, [x(t_-1) for t_ in t])
axes[1].set_xlabel('$t$')
axes[1].set_ylabel('$x(t-1)$')
axes[1].grid(True)
# x(t+1)
axes[2].plot(t, [x(t_+1) for t_ in t])
axes[2].set_xlabel('$t$')
axes[2].set_ylabel('$x(t+1)$')
axes[2].grid(True)
# x(-t)
axes[3].plot(t, [x(-t_) for t_ in t])
axes[3].set_xlabel('$t$')
axes[3].set_ylabel('$x(-t)$')
axes[3].grid(True)
# x(-t+1)
axes[4].plot(t, [x(-t_+1) for t_ in t])
axes[4].set_xlabel('$t$')
axes[4].set_ylabel('$x(-t+1)$')
axes[4].grid(True)
# x(3t/2)
axes[5].plot(t, [x(3*t_/2) for t_ in t])
axes[5].set_xlabel('$t$')
axes[5].set_ylabel('$x(3t/2)$')
axes[5].grid(True)
# x(3t/2+1)
axes[6].plot(t, [x(3*t_/2 +1) for t_ in t])
axes[6].set_xlabel('$t$')
axes[6].set_ylabel('$x(3t/2 +1)$')
axes[6].grid(True)
plt.show()
```

#1.8 Observing a Signal in Frequency Domain

```python
fs = 100# 100−Hz sampling frequency
ts = 1/fs
t = np.arange(-1., 1., ts) # A linearly−spaced array with step ts
fig, axes = plt.subplots(2,1, figsize=(10,10))
f = 2 # 2 Hz
omega0 = 2*np.pi*f
xt = 0.75 + 2.*np.cos(omega0*t) + 1.*np.cos(3*omega0*t)
Xf = np.fft.fft(xt)
freq = np.fft.fftfreq(t.shape[-1], d=ts)
axes[0].plot(t,xt)
axes[0].set_xlabel('$t$ in s')
axes[0].set_ylabel('$x(t)$')
valsubrange = np.concatenate((np.arange(0,21,1), np.arange(-1,-21,-1)))
freqsubrage = np.concatenate((np.arange(0,21,1), np.arange(-1,-21,-1)))
axes[1].stem(freq[freqsubrage], Xf.real[valsubrange]/len(t))
axes[1].set_xlabel('$f$ in Hz')
axes[1].set_ylabel('$\mathfrak{Re}(X(j\omega))$')
plt.xticks(np.arange(-10,11))
plt.show()
```
